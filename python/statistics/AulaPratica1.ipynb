# **`Distribuição Binomial`**


Vamos simular o lançamento de $n$ moedas e a probabilidade de sair $k$ caras. Ou seja, termos $k$ sucessos em $n$ realizações de um experimento.

```python
from random import seed
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import binom
import math

seed(123) # semente do gerador de números aleatórios
n = 100 # número de lançamentos
p = 0.3 # probabilidade de sair cara
P = np.random.binomial(n, p, 1000) 
k = np.arange(0, np.max(P))
count, bins, ignored = plt.hist(P, bins=k, density=True, color='#0504aa',alpha=0.7, 
                                rwidth=0.9)
plt.xlabel('k', fontsize = 15)
plt.ylabel('P(k)',fontsize = 15)

# curva teórica
Pk = binom.pmf(k,n,p)
plt.plot(k, Pk, 'r--', label='Prob. Teórica')
plt.legend()
plt.show(True)
```

# **`Distribuição de Poisson`**
Para gerarmos dados com uma distribuição de Poisson, podemos usar a biblioteca Numpy.


```python
from scipy.stats.distributions import poisson
import numpy as np
import matplotlib.pyplot as plt

seed(123) # semente do gerador de números aleatórios
lbd = 100 # taxa
n = 1000 # numero de pontos extraídos de uma distribuição de Poisson
P = np.random.poisson(lbd, n)
k = np.arange(0, np.max(P))
count, bins, ignored = plt.hist(P, bins=k, density=True, color='#0504aa',alpha=0.7, 
                                rwidth=0.9)
plt.xlabel('k', fontsize = 15)
plt.ylabel('P(k)',fontsize = 15)
# Curva Teorica
Pk = poisson.pmf(k,lbd)
plt.plot(k, Pk, 'r--', label='Prob. Teórica')
plt.legend()
plt.show(True)
```

# **`Distribuição Geométrica`**
Uma variável aleatória X segue uma distribuição geométrica se:

$$P(X=k) = p(1-p)^{k-1} \hspace{0.5cm} k = 0,1,2,\ldots$$ 

Para gerarmos dados com uma distribuição geométrica, podemos usar a biblioteca Numpy.


```python
from scipy.stats import geom
import numpy as np
import matplotlib.pyplot as plt

p = 0.2 # probabilidade de sucesso
n = 10000 # numero de pontos extraídos de uma distribuição de Geométrica
X = np.random.geometric(p, size=n)
k = np.arange(0, np.max(X))
count, bins, ignored = plt.hist(X, bins=k, density=True, color='#0504aa',alpha=0.7, 
                                rwidth=0.9)
plt.xlabel('k', fontsize = 15)
plt.ylabel('P(k)',fontsize = 15)
# Curva Teorica
Pk = geom.pmf(k,p)
plt.plot(k, Pk, 'r--', label='Prob. Teórica')
plt.legend()
plt.show(True)
print("Valor esperado Teórico:", 1/p)
print("Valor esperado pela simulação:", np.sum(X)/len(X))
```

# **`Distribuição Hipergeométrica`**

Considere um conjunto de N objetos, dos quais m são do tipo I e N-m são do tipo II. Para um sorteio de r objetos $(r<N)$, feito ao acaso e sem reposição, defina: X: número de objetos selecionados do tipo I.
Então:
 
 $${\displaystyle P(X=k)={\frac {{\binom {N_1}{k}}{\binom {N+2}{r-k}}}{\binom {N}{r}}}, N = N_1+N_2}$$

Para gerarmos dados com uma distribuição geométrica, podemos usar a biblioteca Numpy.


```python
import numpy as np
import matplotlib.pyplot as plt
seed(123)
N1 = 4 # number de elementos do tipo I
N2 = 21 # numero de elementos do tipo II
r = 5 # numero de objetos selecionados sem reposicao
n = 1000 # numero de pontos extraídos da distribuição hipergeométrica
X = np.random.hypergeometric(N2, N1, r, 1000)
k = np.arange(0, np.max(X) + 5)
print("Valor esperado Teórico:", N2*N1/(N1+N2))
print("Valor esperado pela simulação:", np.sum(X)/len(X))

count, bins, ignored = plt.hist(X, bins=k, density=True, color='#0504aa',alpha=0.7, 
                                rwidth=0.9)
plt.xlabel('k', fontsize = 15)
plt.ylabel('P(k)',fontsize = 15)
plt.show(True)
```
